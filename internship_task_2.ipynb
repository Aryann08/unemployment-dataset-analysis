#VISUALIZING OF DATASET

```python
plt.figure(figsize=(10,6))
plt.style.use("dark_background")
sns.countplot(y=df['Region'])
```

```python
plt.figure(figsize=(10,6))
plt.style.use("dark_background")
sns.heatmap(df.corr(), annot=True)
plt.plot()
```

```python
plt.figure(figsize=(12,6))
plt.style.use("dark_background")
sns.countplot(x=df['Area'])
```

```python
plt.style.use("dark_background")
sns.pairplot(df,hue='Region',palette="rainbow", size=3)
```

```python
sns.lineplot ( data = df , x = "Region" , y = " Estimated Unemployment Rate (%)" )
plt.xticks(rotation=90)
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
fig = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',template='plotly_dark',)
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
fig = px.box(df,x='Region',y=' Estimated Unemployment Rate (%)',template='plotly_dark')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.scatter(df,x='Region',y=' Estimated Unemployment Rate (%)',template='plotly_dark')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.histogram(df,x='Region',y=' Estimated Unemployment Rate (%)',template='plotly_dark')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
